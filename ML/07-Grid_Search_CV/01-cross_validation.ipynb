```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
## CREATE X and y
X = df.drop('sales',axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# SCALE DATA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#cross_val_score
```

```python
# scoring/evaluation options : 
# https://scikit-learn.org/stable/modules/model_evaluation.html
from sklearn.linear_model import Ridge
model = Ridge(alpha=100)
```

```python
from sklearn.model_selection import cross_val_score
help(cross_val_score)
```

```python
scores = cross_val_score(model, # our ridge(alpha=100),
                         X_train,y_train,
                         scoring="neg_mean_squared_error", # score that we like to compare
                         cv=5,# value for K-fold so K = CV = 5 , we decide it
                         verbose=1) # for output result / logging 
```

```python
scores
```

```python
abs(scores.mean())*00.5
```

```python
# we think this is a big error we need to adjust our model 
model = Ridge(alpha=0.02)
```

```python
scores = cross_val_score(model, # our ridge(alpha=0.02),
                         X_train,y_train,
                         scoring="neg_mean_squared_error", # score that we like to compare
                         cv=5,# value for K-fold so K = CV = 5 , we decide it
                         verbose=1) # for output result / logging 
```

```python
scores
```

```python
abs(scores.mean())**0.5
```

```python
# Grid Search 
from sklearn.model_selection import GridSearchCV
help(GridSearchCV)
```

```python
from sklearn.linear_model import ElasticNet
help(ElasticNet)
```

```python
# If we don't have ElasticNetCV and we need to apply Cross validation for a model that has more than one Hyper-parameter 
# we use Grid Search CV technic as next : 

##1 : build base model: 
base_model = ElasticNet()
##2 prepare our dict. for hyper-parameter that needed to calculate in one shot.
grid_param = {"alpha": [0.05,0.02,0.1,1,5,10,50,100],
              "l1_ratio":[.1, .5, .7, .9, .95, .99, 1],
              }
## 3 build grid_model 
grid_model = GridSearchCV(estimator=base_model,
                          param_grid=grid_param,
                          scoring="neg_mean_squared_error",
                          cv=5,
                          verbose=2)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
type(grid_model)
```

```python
grid_model.best_params_
```

```python
abs(grid_model.best_score_)**0.5
```

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
RMSE = mean_squared_error(y_test,y_pred)
RMSE
```

```python
grid_model_r2_score = r2_score(y_test,y_pred)
```
