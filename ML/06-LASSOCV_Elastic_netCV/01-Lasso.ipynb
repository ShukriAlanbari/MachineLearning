```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_convertor = PolynomialFeatures(degree=3 , include_bias=False)
poly_X = polynomial_convertor.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(poly_X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# WE FIT  ON TRAINING DATA NOT TESTING DATA
scaler.fit(X_train)
X_train = scaler.transform(X_train) # or : #X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LassoCV,Lasso
help(LassoCV)
```

```python
help(Lasso)
```

```python
lasso_cv_model = LassoCV(eps=0.001,n_alphas=100)
```

```python
lasso_cv_model.fit(X_train,y_train)
```

```python
# So to not get this warning

lasso_cv_model = LassoCV(eps=0.001,n_alphas=100,max_iter=10_000)
lasso_cv_model.fit(X_train,y_train)
```

```python
lasso_cv_model.alpha_
```

```python
y_pred = lasso_cv_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
MAE = mean_absolute_error(y_true=y_test,y_pred=y_pred)
MAE
```

```python
RMSE = mean_squared_error(y_true=y_test,y_pred=y_pred)**0.5
RMSE
```

```python
lassocv_r2_score = r2_score(y_true=y_test,y_pred=y_pred)
lassocv_r2_score
```

```python
lasso_cv_model.score(X_test,y_test)
```
