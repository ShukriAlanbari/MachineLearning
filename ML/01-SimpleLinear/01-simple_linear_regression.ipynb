Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# read our data set from advertising.csv

```python
df= pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# we shall sum all advertising spend into one column
df["total_spend"] = df["TV"] + df["radio"] + df["newspaper"]
```

```python
df.head()
```

```python
# can we have a linear relation between total_spend and sales columns?

sns.scatterplot(data= df,
                x= "total_spend",
                y= "sales")
```

```python
# how it looks for linear regression between total_spend and sales 

sns.regplot(data=df,
            x= "total_spend",
            y= "sales")
```

```python
# y_hat= beta0 + beta1*x
```

```python
# divide our data frame into independent feature X = df["total_spend"]
# and label Y = df["sales"]

X = df["total_spend"]
y = df["sales"]
```

```python
# to find beta_0 and beta_1 we use polyfit() function from numpy
help(np.polyfit)
```

```python
np.polyfit(X,y,1) # deg = 1 because its a simple regression
```

```python
# output is an array that has [beta_1 ,beta_0]
# y_hat= beta0 + beta1*x

# y_hat = 0.04868788*x + 4.24302822
```

```python
# we predict some future values of sales based on potential spend
potential_spend= np.linspace(0,500,100)
```

```python
potential_spend
```

```python
# y_hat = 0.04868788*x + 4.24302822
predicted_sales_y_hat = 0.04868788*potential_spend + 4.24302822
```

```python
plt.plot(potential_spend,predicted_sales_y_hat)
```

```python
sns.scatterplot(data=df,
                 x="total_spend",
                 y= "sales")
plt.plot(potential_spend,predicted_sales_y_hat,color="red")
```

```python
sns.regplot(data=df, 
            x="total_spend",
            y="sales")
```

```python
def my_prediction(potential_spend:float):
    return 0.04868788*potential_spend + 4.24302822
```

```python
predicted_sales_200 = my_prediction(200)
print(f"if i spend 200k USD my predicted sales is {predicted_sales_200}  K-units")
```
