**#2: From data frame what we shall try to  predict? is it catorigal or continues value?**

```python
# Answer here
# The DataFrame details features and characteristics of houses in the Ames Housing Area,
# the Value That should be predicted would be the [SalePrice] which would be  a continues data 
# which means it is a regression task rather than a classification one  
```

**#3: The label we are trying to predict is the SalePrice column. Separate out the data into X features and y labels**

```python
#Code Here
X = df.drop("SalePrice", axis= 1)
y = df["SalePrice"]
```

**#4: Usually we split data into Training and testing, why we do this?**

```python
#Answer here
# we split the data to ensure that our model can make a general rule for new unseen data ,
# and we use the training set to the train the model the we ecaluate its preformance based on the test set
# this way  would help detect issues in the model predictions like overfitting etc.
```

**#5: Use scikit-learn to split up X and y into a training set and test set. Since we will later be using a Grid Search strategy, set your test proportion to 10%. To get the same data split as the solutions notebook, you can specify random_state = 101**

```python
#code here
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state= 101)
```

**#6: What dose feature Scaling mean? why we scale features?**

```python
#Answer Here
# Feature scaling is a method used in standardizing the range of the independant variables,
# tex: makes them between 0-1 so they can be compared to each other easily 
# as well as speeding up the gradient descent.
```

**#7: The dataset features has a variety of scales and units. For optimal regression performance, scale the X features. Take carefuly note of what to use for .fit() vs what to use for .transform()**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**#8: We will use an Elastic Net model. Create an instance of default ElasticNet model with scikit-learn**

```python
from sklearn.linear_model import ElasticNet

elastic_model = ElasticNet()
```

**#9:  The Elastic Net model has two main parameters, alpha and the L1 ratio. Create a dictionary parameter grid of values for the ElasticNet. using these variables**

param_grid = {'alpha':[0.1,1,5,10,50,100],<br>
              'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}

```python
grid_param = {'alpha':[0.1,1,5,10,50,100],
              'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}
```

**#10: Using scikit-learn create a GridSearchCV object and run a grid search for the best parameters for your model based on your scaled training data.** Just igonre the warnings u get!

```python
from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(estimator = elastic_model,
                          param_grid = grid_param,
                          scoring = "neg_mean_squared_error",
                          cv = 5)
```

```python
grid_model.fit(X_train_scaled, y_train)
```

**#11: Display the best combination of parameters for your model**

```python
grid_model.best_params_
```

**#12: Evaluate your model's performance on the unseen 10% scaled test set. In the solutions notebook we achieved an MAE of $\$$14149 and a RMSE of $\$$20532**

```python
from sklearn.metrics import mean_absolute_error,  mean_squared_error
y_pred = grid_model.predict(X_test_scaled)

elastic_MAE = mean_absolute_error(y_test, y_pred)
elastic_RMSE = abs(grid_model.best_score_)**0.5
```

```python
print("Mean Absolute Error (MAE): {:,.2f}".format(elastic_MAE))
print("Root Mean Squared Error (RMSE): {:,.2f}".format(elastic_RMSE))
```

**#13: compare model.score result with r2_score results, what do think ? and Why?**

```python
#Code Here 
from sklearn.metrics import r2_score
grid_model_r2 = grid_model.best_estimator_.score(X_test_scaled, y_test)

model_r2score = r2_score(y_test, y_pred)
```

```python
#Answer Here
print(f'R^2 from model.score: {grid_model_r2 }')
print(f'R^2 from r2_score: {model_r2score}')
```

**#14: Let's create on the same split of X_test,X_train,y_test,y_train an instance of Linear regression**

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_train)

y_pred_linear = linear_model.predict(X_test_scaled)
```

**#15: Compare MAE,RMSE and R2_score for Elastic_Net and Linear regression, what do you find from these numbers? nad which model you gonna chose and why?**

```python
#Code Here
mae_linear = mean_absolute_error(y_test, y_pred_linear)
rmse_linear = mean_squared_error(y_test, y_pred_linear, squared=False)  # squared=False to get the RMSE
r2_linear = r2_score(y_test, y_pred_linear)

```

```python
#Answer Here
print(f'Linear Regression MAE: {mae_linear:,.2f}')
print(f'Linear Regression RMSE: {rmse_linear:,.2f}')
print(f'Linear Regression R^2: {r2_linear:.5f}')
print ("*"*50)
print("Mean Absolute Error (MAE): {:,.2f}".format(elastic_MAE))
print("Root Mean Squared Error (RMSE): {:,.2f}".format(elastic_RMSE))
print(f'R^2 from model.score: {grid_model_r2 }')
```

**Advanced 16: using cross_val_score from model_selection into SK learn to use Cross Validation of K-fold = 10 to get the scores of each k-fold and print the average for the Elastic_Net model and Linear regression model**

```python
from sklearn.model_selection import  cross_val_score 
elastic_net_scores = cross_val_score(grid_model.best_estimator_,
                         X_train_scaled,
                         y_train,
                         scoring= "neg_mean_squared_error",
                         cv = 10)

linear_model_scores = cross_val_score(linear_model,
                                 X_train_scaled,
                                 y_train,
                                 scoring= "neg_mean_squared_error",
                                 cv = 10)
elastic_net_rmse_scores = np.sqrt(-elastic_net_scores)
linear_model_rmse_scores = np.sqrt(-linear_model_scores)

print("ElasticNet RMSE scores for each fold:")
for fold, score in enumerate(elastic_net_rmse_scores, 1):
    print(f"Fold {fold}: {score:.2f}")
print(f"\nElasticNet Average RMSE: {elastic_net_rmse_scores.mean():.2f}\n")

print(f"{"*"*50}\n")

print("Linear Regression RMSE scores for each fold:")
for fold, score in enumerate(linear_model_rmse_scores, 1):
    print(f"Fold {fold}: {score:.2f}")
print(f"\nLinear Regression Average RMSE: {linear_model_rmse_scores.mean():.2f}")
```
